```python
city=data.groupby(['City'])[['Sales','Discount','Profit']].mean()
```

```python
city=city.sort_values('Profit')
```

```python
city.head(10)
```

```python
# for high profile analysis

city['Profit'].tail(30).plot(kind='bar',figsize=(15,5),color='Green')
plt.title('City Wise Sales,Discount and Profit Analysis')
plt.show()
```

```python
# for low profit analysis
city['Profit'].head(30).plot(kind='bar',figsize=(15,5),color='Blue')
plt.title('City Wise Sales,Discount and Profit Analysis')
plt.show()
```

```python
quantity = data.groupby(['Quantity'])[['Sales','Discount','Profit']]
```

```python
quantity = data.groupby(['Quantity'])[['Sales','Discount','Profit']].mean()
```

```python
quantity.head(10)
```

```python
quantity.plot.pie(subplots=True,autopct='%1.1f%%',figsize=(25,25),
                 pctdistance=0.05,startangle=90,shadow=True,labels=quantity.index)
plt.title('Quantity wise Sales,Discount and Profit Analysis',fontsize=20)
plt.show()
```

```python
cat=data.groupby(['Quantity'])[['Sales','Discount','Profit']]
```

```python
cat=data.groupby(['Quantity'])[['Sales','Discount','Profit']].mean()
```

```python
cat
```

```python
cat.plot.pie(subplots=True,autopct='%1.1f%%',figsize=(25,25),
                 pctdistance=0.05,startangle=90,shadow=True,labels=cat.index)
plt.title('Quantity wise Sales,Discount and Profit Analysis',fontsize=20)
plt.show()
```

```python
sub_cat=data.groupby(['Sub-Category'])[['Sales','Discount','Profit']]
```

```python
sub_cat=data.groupby(['Sub-Category'])[['Sales','Discount','Profit']].mean()
```

```python
sub_cat.head(10)
```

```python
plt.figure(figsize=(25,25))
plt.pie(sub_cat['Sales'],labels=sub_cat.index,autopct='%1.1f%%')
plt.title('sub-category wise sales,profit and discount analysis',fontsize=20)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(25,25))
plt.pie(sub_cat['Discount'],labels=sub_cat.index,autopct='%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis',fontsize=20)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
sub_cat.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',figsize=(20,10),label=
                                                       ['Average Sales Price($)','Profit(%)'])
plt.title('Sub-Category Wise Profit Analysis',fontsize=15)
plt.legend()
plt.show()
```
